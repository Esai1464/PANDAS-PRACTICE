LOADING A DATASET

```python
import pandas as pd
data=pd.read_csv('Product_cleaned.csv')
```

CHECKING NULL VALUES

```python
print(data['Units Sold'].isnull().sum())
print(data['Units Sold'].value_counts(dropna=False).to_string())
```

FINDING DATA TYPES AND CONVERTING INTO CORRECT FORM

```python
print(data['Units Sold'].dtype)
data['Units Sold']=pd.to_numeric(data['Units Sold'],errors='coerce')
print(data['Units Sold'].dtype)
```

CHECKING NEGATIVE VALUES

```python
print(data[data['Units Sold'] < 0])
```

```python
print(data['Units Sold'].describe())
```

FINDING MEDIAN

```python
MED= data['Units Sold'].median()
print(MED)
```

FILLING NULL VALUES WITH MEDIAN BECAUSE IT CONTAINS OUTLIER

```python
data['Units Sold']=data['Units Sold'].fillna(MED)
print(data['Units Sold'].isnull().sum())
```

```python
print(data['Units Sold'])
```

```python
print(data['Units Sold'].value_counts(dropna=False).to_string())
```

REPLACING WRONG VALUES TO MEDIAN

```python

wrong_val=[142730.33,147209.07,169095.82,5482.77,153159.37,152440.84]
data['Units Sold']=data['Units Sold'].replace(wrong_val,MED)
```

```python
print(data[data['Units Sold'].isin(wrong_val)].to_string())
```

```python
print(data['Units Sold'].value_counts(dropna=False).to_string())
```

EXPORTING CSV FILE

```python
from google.colab import files
data.to_csv('Unitsold_cleaned.csv',index=False)
files.download('Unitsold_cleaned.csv')
```
