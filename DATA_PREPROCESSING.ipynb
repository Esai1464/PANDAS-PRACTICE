LOAD A DATASET

```python
import pandas as pd
data=pd.read_csv('nigeria_messy_sales_dataset.csv')
print(data.to_string())
```

CONVERT CUSTOMER NAME COLUMN INTO LOWER CASE


```python
data['Customer Name']=data['Customer Name'].str.lower().str.strip()
print(data)
```

COUNTING VALUES FOR CUSTOMER NAME


```python
print(data['Customer Name'].value_counts(dropna=False).to_string())
```

CHECKING IF ANY . AVAILABLE IN CUSTOMER NAME


```python
print(data[data['Customer Name'] == '.'].sum())
```

CHECK IF ANY WRONG NAMES AVAILABLE IN CUSTOMER NAME

```python
wrong_names = ['plateau','niger','delta','enugu','kano','lagos','abuja']
print(data[data['Customer Name'].isin(wrong_names)].to_string())
```

FILLING NULL VALUES WITH UNKNOWN NAME

```python
data['Customer Name']=data['Customer Name'].fillna('Unkown Name')
print(data['Customer Name'])
```

CORRECTING UNKOWN TO UNKNOWN

```python
data['Customer Name']=data['Customer Name'].replace(wrong_names,'Unknown Name')
print(data['Customer Name'].to_string())

data['Customer Name']=data['Customer Name'].replace('unkown name','Unknown Name')
print(data['Customer Name'].to_string())
```

PRINTING A DATASET


```python
print(data.to_string())
```

```python
print(data[data['Customer Name'].duplicated()])
```

CHECKING NULL VALUES


```python
print(data['Customer Name'].isnull().sum())
```
