CLEANING STATE COLUMN

```python
import pandas as pd
data=pd.read_csv('cleaned_data.csv')
```

```python
data['State']=data['State'].str.title()
data['State']=data['State'].str.strip()
data['State']=data['State'].str.replace(r'\s+',' ',regex=True)

```
