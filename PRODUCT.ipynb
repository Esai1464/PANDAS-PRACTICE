LOADING A DATASET (THAT CONTAINS CLEANED COLUMNS OF CUSTOMER NAME AND STATE)

```python
import pandas as pd
data=pd.read_csv('state_cleaned.csv')
```

PRINTING THE PRODUCT COLUMN

```python
data['Product']=data['Product'].str.title()
print(data['Product'])
```

```python
print(data['Product'].unique())
print(data['Product'].value_counts(dropna=False))
```

MANAGING WHITE SPACES

```python
data['Product']=data['Product'].str.strip()
data['Product']=data['Product'].str.replace(r'\s+',' ',regex=True)
```

```python
print(data['Product'].to_string())
```

CHECKING WRONG VALUES

```python
wrong_num = ['5578','4567!!']
chec=data['Product'].isin(wrong_num)
print(data[chec].to_string())
```

REPLACING WRONG VALUE WITH OTHERS

```python
data['Product']=data['Product'].replace(wrong_num,'Others')
print(data.loc[[486,554],['Product']])
```

```python
print(data['Product'].to_string())
```
