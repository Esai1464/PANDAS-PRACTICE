```python
import pandas as pd
data=pd.read_csv('sale_date_cleaned.csv')
```

```python
print(data['Sales Channel'].value_counts(dropna=False))
```

```python
data['Sales Channel']=data['Sales Channel'].fillna('Direct')
```

```python
print(data['Sales Channel'].value_counts(dropna=False))
```

```python
wrong_val=['039b058c-2faa-4e0b-b7f6-0cdf41dc1c60',
'4b86b5a1-ca6f-4ff8-964c-fbd2bc92fd81',
'317304b5-c1dd-484a-aca5-8d95915a1c30',
'a383889a-db2c-4ac8-9c67-83932ced3c0d',
'6d7df5d0-d3b5-4af7-a7ce-378fe5a4983b',
'669d01ff-1634-425b-93f3-007383e96ef4']
data['Sales Channel']=data['Sales Channel'].replace(wrong_val,'Direct')
```

```python
print(data['Sales Channel'].value_counts(dropna=False))
```

```python
print(data['Sales Channel'].info())
print(data['Sales Channel'].dtype)
```

```python
print(data['Order ID'].value_counts(dropna=False))
```

```python
print(data['Order ID'].isnull().sum())
```

```python
duplicate_orders = data[data['Order ID'].duplicated(keep=False)]
print(duplicate_orders.to_string())
```

```python
data = data.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
duplicate_orders = data[data['Order ID'].duplicated(keep=False)]
print(duplicate_orders.to_string())
```

```python
data = data.drop_duplicates(subset=['Order ID'], keep='first')
print(data)
```

```python
duplicate_orders = data[data['Order ID'].duplicated(keep=False)]
print(duplicate_orders.to_string())
```
