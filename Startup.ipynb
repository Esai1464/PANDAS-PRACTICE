#LOAD DATASET

```python
import pandas as pd
data=pd.read_csv('nigeria_messy_sales_dataset.csv')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

# Level 2 - Selecting & Filtering

  ## loc vs iloc

```python
print(data.iloc[9:20].to_string())
print()
print(data.loc[5:15].to_string())
print()
print(data.iloc[-5:-1].to_string())

print()

print(data.loc[:,['Customer Name', 'Total Sale']])
```

 ## Single Condition

```python
data['State']=data['State'].str.lower()
lagstate=data[data['State']=='lagos']
print(len(lagstate))
```

```python
data['Total Sale']=pd.to_numeric(data['Total Sale'],errors='coerce')
totsal=data[data['Total Sale']>100000]
print(totsal.to_string())
print(len(totsal))
```

```python
print(data.info())
```

```python
data['Unit Sale']=pd.to_numeric(data['Units Sold'],errors='coerce')
unitsal= data[data['Unit Sale']<5]['Customer Name']
print(unitsal)
```

```python
a=data[~data['Sales Channel'].isin(['Online'])]
print(a.to_string())
```

## Multiple Conditions

```python
data['State']=data['State'].str.lower()
st=data[(data['State']=='lagos') & (data['Total Sale']>50000)]
print(st.to_string())
```

##query

```python
result=data.query("State=='lagos'")
print(result)
```

```python
data['Total Sale'] = pd.to_numeric(data['Total Sale'], errors='coerce')
res=data.query('`Total Sale` > 50000 and State == "kano"')
print(res.to_string())
```

```python
data['Units Sold']=pd.to_numeric(data['Units Sold'],errors='coerce')
data['Sales Channel']=data['Sales Channel'].str.lower()
res=data.query('`Units Sold` > 10 and `Sales Channel` == "retail"')
print(len(res))

```
